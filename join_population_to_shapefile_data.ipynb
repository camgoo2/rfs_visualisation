# Joining population data of LGA's to LGA polygon data



```python
import pandas as pd
import shapefile as shp
import geopandas as gpd
```

```python
#read the LGA polygon shapefile
#as per geo_df_pop.to_file('1259030001_lga11aaust_shape/LGA11aAust.shp') code futher below, 
#if we open up the file again it will now contain the pop data as we have saved it to it.
geo_df = gpd.read_file('1259030001_lga11aaust_shape/LGA11aAust.shp')

#view LGA polygon shapefile
geo_df
```

```python
#read population csv file
population_df = pd.read_csv('2011_BCP_LGA_for_AUST_short-header/2011 Census BCP Local Government Areas for AUST/AUST/2011Census_B01_AUST_LGA_short.csv')
```

```python
#Rename the column of the region_id column to be the same as our geo_df dataframe "LGA_CODE11"
population_df = population_df.rename(columns={'region_id':'LGA_CODE11'})
```

```python
#remove 'LGA' from the  data'LGA_CODE11' column to be consistent with the LGA_CODE11 data in geo_df
#need the regex agrument here otherwise it won't replace
population_df = rename_df.replace('LGA','', regex=True)
```

```python
population_df
```

```python
#isin method
#remove row based on condition (population_df 577 rows versus 565 rows, we want these to remove any rows that are surplus)
population_df = population_df[population_df['LGA_CODE11'].isin(geo_df['LGA_CODE11'])]
```

```python
population_df
```

```python
# population_df reduced from 577 rows × 109 columns to 565 rows × 109 columns"
```

```python
#extract the total population column
extracted_population_col = population_df["Tot_P_P"]

#add total population to geo_df
geo_df_pop = geo_df.join(extracted_population_col)
```

```python
geo_df_pop
```

```python
#this has added our population data to our file, 
geo_df_pop.to_file('1259030001_lga11aaust_shape/LGA11aAust.shp')
```

```python
def create_distinct_data_map(df,folium):
    fire_location = df[["latitude", "longitude"]]
    map = folium.Map(location=[fire_location.latitude.mean(), fire_location.longitude.mean()],
                 zoom_start=5, control_scale=True)
    for index, location_info in fire_location.iterrows():
        folium.Marker([location_info["latitude"], location_info["longitude"]]
                      ,icon=folium.Icon(color="orange",icon = "fire")).add_to(map)
    #have used https://mapshaper.org/ to reduce the file size of polygons. I am now using 
    #reduced file LGA11aAust.json rather than LGA11aAust.geojson 
    folium.GeoJson('LGA11aAust.json').add_to(map)
    
    return map
```
