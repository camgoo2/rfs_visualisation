```python
import xml.etree.ElementTree as ET
mytree = ET.parse('simple.xml') #Saved file to computer,where we are in the working directory , so we can use rather having to type the whole file path 
myroot = mytree.getroot()
```

```python
for node in myroot: 
    f_name = node.find("name")
    f_price = node.find("price").text
    f_description = node.find("description").text
    f_calories = node.find("calories").text
```

```python
df_cols = ["name", "price", "description", "calories"]
rows = []

for node in myroot: 
    f_name= node.find("name")
    f_price = node.find("price").text if node is not None else None
    f_description = node.find("description").text if node is not None else None
    f_calories = node.find("calories").text if node is not None else None
    
    rows.append({"name": f_name, "price": f_price, 
                 "description": f_description, "calories": f_calories}) #nesting a dictionary in a list?

out_df = pd.DataFrame(rows, columns = df_cols)
```
